```python
null_proportion = df.isnull().mean()

columns_with_mostly_nulls = null_proportion[null_proportion > 0.8].index.tolist()

print("Columnas con mÃ¡s del 90% de valores nulos:")
print(len(columns_with_mostly_nulls))
```

```python
# cuantos de los registros pertenecen a miembros crea
df["extra_crea"] = df["extra_crea"].fillna("no")
extra_crea = df.groupby(["extra_crea"])["id"].count().reset_index()

plt.barh(extra_crea["extra_crea"], extra_crea["id"], color='skyblue')
plt.xlabel("Cantidad")
plt.ylabel('Miembro CREA')
plt.title('Horizontal Bar Chart')
plt.show()

extra_crea = extra_crea.fillna('no')
extra_crea
```

```python
mat_org_prov = df.groupby(["provincia"])["materia_organica_porc"].mean()
mat_org_prov = mat_org_prov.reset_index().sort_values("materia_organica_porc", ascending  = False)

plt.xticks(rotation=45, ha='right')  

sns.barplot(x="provincia", y="materia_organica_porc", data = mat_org_prov)

# Add labels and title
plt.xlabel('Provincia')
plt.ylabel('Media de mat org')
plt.title('Distribucion de datos por prov')

# Show plot
plt.show()
```

```python
df['fecha_cosecha'] = pd.to_datetime(df['fecha_cosecha'])

# Extraer los aÃ±os de la columna 'fecha_cosecha'
df['anio_cosecha'] = df['fecha_cosecha'].dt.year

df['anio_cosecha'].sort_values(ascending=True)

cantidad_cosechas_anio = df.groupby(["anio_cosecha"])["id"].count().reset_index()

# grafico 

plt.xticks(rotation=45, ha='right')  

sns.barplot(x="anio_cosecha", y="id", data = cantidad_cosechas_anio)

# Add labels and title
plt.xlabel('Ano de cosechado')
plt.title('Distribucion de datos por prov')

# Show plot
plt.show()
cantidad_cosechas_anio
```
